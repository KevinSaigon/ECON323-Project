```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("salaries.csv")
```

# An example of how dataset use in this project looks like

```python
display(data.head(10))
```

# Visualization of annual base pay versus seniority 
#### Using Seaborn histogram to visualize annual_base_pay because there are more than 2 levels of seniority

```python
fig, ax = plt.subplots()
sns.histplot(x='annual_base_pay', data=data.head(100), ax=ax, kde=True, hue='job_title_rank')

plt.show()
```

# Visualization of annual base pay versus different job title
#### Likewise, using Seaborn histogram to visualize annual_base_pay because there are more than 2 different types of job title

```python
fig, ax = plt.subplots()
sns.histplot(x='annual_base_pay', data=data.head(100), ax=ax, kde=True, hue='job_title_category')
plt.show()
```

# Visualization of annual base pay versus different country
#### Likewise, using Seaborn histogram to visualize annual_base_pay because there are more than 2 different countries

```python
fig, ax = plt.subplots()
sns.histplot(x='annual_base_pay', data=data.head(100), ax=ax, kde=True, hue='location_country')
plt.show()
```

# Visualization of annual base pay versus experience years (employer) 
#### Using Bar Chart to visualize the relationship between annual_base_pay and experience year

```python
fig, ax = plt.subplots()

 
# Bar chart with day against tip
plt.bar(data.head(100)['employer_experience_years'], data.head(100)['annual_base_pay'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('employer_experience_years')
plt.ylabel('annual_base_pay')
 
# Adding the legends
plt.show()
```

# Visualization of annual base pay versus experience years (employer) 
#### Using Scatter Plot to visualize the relationship between signing)_bonus and annual_base_pay

```python
plt.scatter(data.head(100)['signing_bonus'], data.head(100)['annual_base_pay'])
plt.title("Scatter Plot")

plt.xlabel('signing_bonus')
plt.ylabel('annual_base_pay')
```
